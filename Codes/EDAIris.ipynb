```python
import numpy as np
```

```python
q1 = np.percentile(df['SepalWidthCm'],25)
q3 = np.percentile(df['SepalWidthCm'],75)
```

```python
IQR = q3-q1
```

```python
IQR
```

```python
ll = q1-(1.5*IQR)
ul = q3 + (1.5*IQR)
```

```python
up = df[df['SepalWidthCm']>ul].index.tolist()
low = df[df['SepalWidthCm']<ll].index.tolist()
```

```python
bad_indices =  list(set(up+low))
```

```python
drop = True
```

```python
if drop :
    df.drop(bad_indices ,inplace=True , errors='ignore')
```
